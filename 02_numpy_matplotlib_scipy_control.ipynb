```python
# Followings are matplotlib tips
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
# オブジェクト指向のグラフ
"""
Containerの階層構造
Figure
 - Axes
  - Axis
   - (XAxis, YAxis)
     - Tick
"""
fig, ax = plt.subplots()  #Figure と Axes　オブジェクトの作成
ax.plot(x, y) # Axesオブジェクトの中にグラフを作成
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.grid()
plt.show()
```

```python
fig, ax = plt.subplots(2, 1)

x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
z = np.cos(x)
w = y + z

# 一つのグラフを作成
ax[0].plot(x, y, ls='-', label='sin', c='k')
ax[0].plot(x, z, ls='-.', label='cos', c='k')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y, z')
ax[0].set_xlim(0, 4*np.pi)
ax[0].grid()
ax[0].legend()

#二つのグラフを作成
ax[1].plot(x, w, color='k', marker='.')
ax[1].set_xlabel('x')
ax[1].set_ylabel('w')
ax[1].set_xlim(0, 4*np.pi)
ax[1].grid(ls=':')

fig.tight_layout() #グラフがかさらない
```

#### 微分工程式

$\dot{y} = -\frac{1}{5}y(t) + \frac{1}{5}u(t)$

入力$u(t)$を

$u(t) = 0 (t<10); 1 (t=>10)$

```python
# 微分工程式の数値積分
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt

#微分工程式を定義する
def system(y, t):
    if t < 10.0:
        u = 0.0
    else:
        u = 1.0
    dydt = (-y + u) / 5.0
    return dydt

# 初期値を時間を設定して、微分工程式を解く
y0 = 0.5
t = np.arange(0, 40, 0.04)
y = odeint(system, y0, t)

# グラフを描画
fig, ax = plt.subplots()
ax.plot(t, y , label='y')
ax.plot(t, 1*(t>10), ls='--', label='u')
ax.set_xlabel('t')
ax.set_ylabel('y,u')
ax.legend(loc='best')
ax.grid(ls=':')
```
